#### Formula for Covariance

```python
# Covariance
import numpy as np
import pandas as pd
import seaborn as sns

x = np.array([[1,-1,2,3,5], [3,-2,4,6,9]])
print("Covariance matrix of x:\n", np.cov(x,))

```

### Formula for Pearson's Correlation

```python
df = pd.DataFrame([(1,3),(-1,-2),(2,4),(3,6),(5,9),(8,10)],columns=['x','y'])
pd.DataFrame(x)
df.corr(method='pearson')
# df.std() # x-> 3.16 ,y-> 4.38
# df.cov() -> 13.2
# 13.2/(3.16*4.38) #-> 95.3%
```

```python
# Pearsons Correlation Coefficient
# Used for variables with linear relationship

temp = [(14.2,215),(16.4,325),(11.9,185),(15.2,332),(18.5,406),(22.1,522),(19.4,412),(25.1,614),(23.4,544),(18.1,421),(22.6,445),(17.2,408)]
df = pd.DataFrame(temp,columns=['Temperature','Ice Cream Sales'])
print(df.corr(method='pearson'))

sns.lineplot(x=df['Temperature'],y=df['Ice Cream Sales'])
```

```python
# Spearman Correlation Coefficient 
# Used for variables with non linear relationship

import pandas as pd
import seaborn as sns

temp = [(14.2,215),(16.4,325),(11.9,185),(15.2,332),(18.5,406),(22.1,522),(19.4,412),(25.1,614),(23.4,544),(18.1,421),(22.6,445),(17.2,408)]
df = pd.DataFrame(temp,columns=['Temperature','Ice Cream Sales'])
print(df.corr(method='spearman'))
sns.lineplot(x=df['Temperature'],y=df['Ice Cream Sales'])
```
