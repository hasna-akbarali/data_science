### different types of graph

1. bar graph - Suitable for displaying comparison between categorical data and discrete data. 
2. histogram - Suitable for visualizing the frequency distribution of a single variable.
3. pie chart - helps organize and show data as a percentage of a whole for categorical variables.
4. scatter plot -  scatter plots to visualize the relationship between two continuous variables.
5. line chart - suitable for showing trends, how one variable changes over time


```python
# 1. Bar Graph  
# x - categorical
# y - numerical

products = ['Milk', 'Bread', 'Eggs', 'Curd']
sales = [100, 150, 120,140]
plt.figure(figsize=(2,2))
plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales (units)')
plt.title('Sales Performance of Grocery Products')
plt.show()
```

```python
# Histogram - 
# x - data
# y - freq / count

scores = [85, 72, 90, 65, 78, 82, 95, 88, 75, 80]

plt.hist(scores, bins=5,edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')
plt.show()
```

```python
# Pie Chart

exp_values = [1400, 600, 300, 410, 250]
exp_labels = ["Rent", 'Food' ,'Bills', 'Car' , 'Other utilties']

plt.pie(exp_values, labels=exp_labels, autopct='%0.1f%%')

plt.savefig("pie_chat.pdf")
```

#### Creating subplots

```python
import pandas as pd
df = pd.read_csv('Datasets/avocado.csv')
df.head(2)
```

```python
plt.figure(figsize=(3,3))
plt.bar('Type','AveragePrice',data=df)
```

```python
df['Region']
```

```python
fig, axs = plt.subplots(3, figsize=(10, 15))  # Adjusting figsize for better visualization

# First subplot (bar chart)
axs[0].bar('Type','AveragePrice',data=df)
axs[0].set_xlabel('Products')  
axs[0].set_ylabel('Average Price') 
axs[0].set_title('Type of Avocado vs Average Price')

# Second subplot (scatter plot)
axs[1].scatter(df['Total Volume'], df['AveragePrice'])
axs[1].set_xlabel('Total Volume')  
axs[1].set_ylabel('Average Price')  
axs[1].set_title('Total Volume vs Average Price')

# Third subplot (pie chart)
region_counts = df['Region'].value_counts()
axs[2].pie(region_counts, labels=region_counts.index, autopct='%0.2f%%')
axs[2].set_title('Distribution of Avocado Sales by Region')  # Title adjustment

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# Scatterplot

sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.histplot(data=tips, x='total_bill', bins=20)
```

```python
sns.barplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
# sns.relplot(x='total_bill', y='tip', data=tips)
sns.relplot(x=tips['total_bill'] , y=tips['tip']) #basic
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue='smoker') # color
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue='smoker', style='time') #shapes
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue='smoker', style='time', size='sex')
```
