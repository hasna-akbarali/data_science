Indexing using loc
- you can slice both rows as well as columns
- custom indexing
- syntax - dataframe.loc[start:stop:step (rows),start:stop:step (col)]
- Zero indexing and custom indexing start and stop is included

```python
df = bike_data.head(5)
# df.loc[0:2,'tripduration':'stoptime']
df.index = ['a','b','c','d','e']
df.loc['a':'b','tripduration':'stoptime']
```

##### Add/Update/Delete column 
 - direct -> df[new_column_name] = 'value' column exist then overwrites
 - insert -> df.insert(position,column name, values)
 - update value of particular column -> df.loc[row,col] = value
 - update column name -> df.rename(columns={'old_name': 'new_name'},inplace=True)
 - delete a column -> df.drop(columns='tripduration')

```python
data = pd.read_csv("datasets/international_org.csv")
```

```python
# direct -> df[new_column_name] = 'value' column exist then overwrites
data['Branch Offfice'] = 'Bengaluru'
data
```

```python
# insert -> df.insert(position,column name, values)
# List of the values being passed needs to be equal the number of rows in the dataframe.
data.insert(1,'Departments',['Accounts','Analytics','Sales','Operations','HR','Management'])
```

```python
data
```

```python
# update value of particular column -> df.loc[row,col] = value
data.loc[3,'Headquater'] = 'Bengaluru'
data
```

```python
data.loc[3,'Department'] = 'Development'
data
```

```python
# update column name -> df.rename(columns={'old_name': 'new_name'},inplace=True)
# First make a copy of the dataframe, then makes changes to the new copy
data.rename(columns={'Headquater':'HeadQuaters'},inplace=True)
data
```

```python
# delete a column -> df.drop(columns='tripduration')
data.drop(columns='Branch Office',inplace=True)
data
```

##### Concatenation and Merging
    - Concatenation - rows 0
                    - col 1

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
pd.concat([india_weather, us_weather]) # Default axis = 0 ie rows
```

```python
pd.concat([india_weather, us_weather],axis=1)
```

```python
## merge

# - works with two dataframe
# - we should have same columns to merge
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
# Merging will never happen if we dont have a common column name

df3 = df2.drop(columns='city')
pd.merge(df1, df3, how='outer')
```

```python
pd.merge(df1, df2, how='outer')
```

##### Handling Missing Values
     - fill with value
     - drop all null values

```python
df = pd.read_csv("datasets/weather_data_nan.csv")
df
```

```python
# Fill with any value
df.fillna(0)
```

```python
# different values for diff columns   (inplace=True)

df.fillna({
    'temperature':df['temperature'].mean(),
    'windspeed':20,
    'event':"no event"
})
```

```python
# Dropping all the rows with null values (inplace=True)
#  axis = 0 / rows
# axis = 1 / col
df.dropna(axis=0)
```

#### Duplicates in Dataset

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Drop if any duplicates are present
df.drop_duplicates(inplace=True)
```
